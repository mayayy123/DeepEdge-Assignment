```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def data_augmentation(image):
    # Randomly flip the image horizontally
    image = tf.image.random_flip_left_right(image)

    # Randomly flip the image vertically
    image = tf.image.random_flip_up_down(image)

    return image
```

```python
# Create a list of all image file paths in the data_new directory
image_files = tf.io.gfile.glob('/kaggle/working/data_new/*.png')
image_files = sorted(image_files)
```

```python
# Create a list of all image file paths in the data_new directory
image_files = tf.io.gfile.glob('/kaggle/working/data_new/*.png')
image_files = sorted(image_files)
# Create a dataset from the image file paths
dataset_images = tf.data.Dataset.from_tensor_slices(image_files)

# Apply data augmentation to each image
dataset_images = dataset_images.map(
    lambda x: tf.io.read_file(x),
    num_parallel_calls=tf.data.experimental.AUTOTUNE)
dataset_images = dataset_images.map(
    lambda x: tf.image.decode_png(x, channels=3),
    num_parallel_calls=tf.data.experimental.AUTOTUNE)
#dataset_images = dataset_images.map(data_augmentation, num_parallel_calls=tf.data.experimental.AUTOTUNE)   no augmentation because it changes coordinate positions

# Create a dataset from the CSV data
dataset_csv = tf.data.Dataset.from_tensor_slices(df.values)

# Combine the image and CSV datasets into a single dataset
dataset = tf.data.Dataset.zip((dataset_images, dataset_csv))

# Batch the dataset
dataset = dataset.batch(1)

# Prefetch the data
dataset = dataset.prefetch(buffer_size=tf.data.experimental.AUTOTUNE)

# Print the number of batches in the dataset
print(len(list(iter(dataset))))

```

```python
#Sample data from dataset
# Get the first batch of data
images, labels = next(iter(dataset))

# Print the first 5 images and their corresponding labels
for i in range(1):
    plt.imshow(images[i].numpy())
    plt.scatter(labels[i][0].numpy(),labels[i][1].numpy())
    plt.title(f"Label: {labels[i]}")
    plt.axis('off')

# Display the plot
plt.show()
```

```python
from tensorflow import keras
from tensorflow.keras import layers

# Define the optimized model
model = keras.Sequential([
    layers.InputLayer(input_shape=(256, 256, 3)),
    layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(256, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(512, (3, 3), activation='relu', padding='same'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dropout(0.5),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(256, activation='relu'),
    layers.Dense(2)
])

# Compile the optimized model
model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])

# Display the optimized model summary
model.summary()

# Train the optimized model
model.fit(dataset, epochs=120)
```

```python
# Save the model as "model.keras"
model.save('model.h5')
```
